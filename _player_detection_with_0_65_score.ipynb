```python
data_yaml = """
train: /content/data/train/images
val: /content/data/valid/images

nc: 3
names: ["class0", "class1", "class2"]
"""
with open("data.yaml", "w") as f:
    f.write(data_yaml)
```

```python
from ultralytics import YOLO


model = YOLO('yolov8n.pt')


model.train(
    data="data.yaml",
    epochs=10,
    imgsz=416,
    batch=8
)


model.save('best.pt')
```

```python
import os
import pandas as pd
from ultralytics import YOLO

# Load model
model = YOLO("best.pt")

# Load test image names
test_csv_path = "/content/data/images_test.csv"
test_df = pd.read_csv(test_csv_path)
test_image_ids = test_df["image_id"].tolist()

test_image_folder = "/content/data/valid/images"

submission_rows = []

for img_id in test_image_ids:
    img_path = os.path.join(test_image_folder, f"{img_id}.jpg")

    if not os.path.exists(img_path):
        print(f" Missing image: {img_path}")
        submission_rows.append({"image_id": img_id, "prediction_string": ""})
        continue

    result = model.predict(source=img_path, conf=0.25, save=False, verbose=False)[0]

    prediction_str = ""
    if result.boxes is not None and result.boxes.data.shape[0] > 0:
        for box in result.boxes.data.tolist():
            x1, y1, x2, y2, conf, cls = box
            prediction_str += f"Player {conf:.4f} {int(x1)} {int(y1)} {int(x2)} {int(y2)} "
        prediction_str = prediction_str.strip()

    submission_rows.append({
        "image_id": img_id,
        "prediction_string": prediction_str
    })

# Save final submission
submission_df = pd.DataFrame(submission_rows)
submission_df.to_csv("submission.csv", index=False)
print("Final submission saved with", len(submission_df), "rows.")
```
